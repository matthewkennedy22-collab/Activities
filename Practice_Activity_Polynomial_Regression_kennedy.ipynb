```python

penguins_dummies['bill_leng_2'] = penguins_dummies['bill_length_mm']**2
penguins_dummies['bill_leng_3'] = penguins_dummies['bill_length_mm']**3
penguins_dummies['bill_leng_4'] = penguins_dummies['bill_length_mm']**4
penguins_dummies['bill_leng_5'] = penguins_dummies['bill_length_mm']**5
penguins_dummies['bill_leng_6'] = penguins_dummies['bill_length_mm']**6
penguins_dummies['bill_leng_7'] = penguins_dummies['bill_length_mm']**7
penguins_dummies['bill_leng_8'] = penguins_dummies['bill_length_mm']**8
penguins_dummies['bill_leng_9'] = penguins_dummies['bill_length_mm']**9
penguins_dummies['bill_leng_10'] = penguins_dummies['bill_length_mm']**10
penguins_dummies





```

```python
penguins2 = penguins_dummies.dropna()
penguins2
```

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split

y = penguins2[['bill_depth_mm']]
X = penguins2[['bill_length_mm']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) # Added random_state for reproducibility
```

```python
idx_tr = X_train.index
idx_te = X_test.index
```

1. Linear Model

```python
X_lin_tr  = penguins2.loc[idx_tr, ['bill_length_mm']]
X_lin_te  = penguins2.loc[idx_te, ['bill_length_mm']]
y_train_lin = penguins2.loc[idx_tr, ['bill_depth_mm']]
y_test_lin = penguins2.loc[idx_te, ['bill_depth_mm']]
```

```python
# Linear Model
lin_model = LinearRegression()
lin_model.fit(X_lin_tr, y_train_lin)
y_lin_pred = lin_model.predict(X_lin_te)
```

2. Quadratic Model

```python
X_sq_tr   = penguins2.loc[idx_tr, ['bill_length_mm', 'bill_leng_2']]
X_sq_te   = penguins2.loc[idx_te, ['bill_length_mm', 'bill_leng_2']]
y_train_sq = penguins2.loc[idx_tr, ['bill_depth_mm']]
y_test_sq = penguins2.loc[idx_te, ['bill_depth_mm']]
```

```python
# Quadratic Model
sq_model = LinearRegression()
sq_model.fit(X_sq_tr, y_train_sq)
y_sq_pred = sq_model.predict(X_sq_te)
```

3. Cubic Model

```python
X_cub_tr  = penguins2.loc[idx_tr, ['bill_length_mm', 'bill_leng_2', 'bill_leng_3']]
X_cub_te  = penguins2.loc[idx_te, ['bill_length_mm', 'bill_leng_2', 'bill_leng_3']]
y_train_cub = penguins2.loc[idx_tr, ['bill_depth_mm']]
y_test_cub = penguins2.loc[idx_te, ['bill_depth_mm']]
```

```python
# Cubic Model
cub_model = LinearRegression()
cub_model.fit(X_cub_tr, y_train_cub)
y_cub_pred = cub_model.predict(X_cub_te)
```

4. Degree 10 Polynomial Model

```python
X_deg10_tr = penguins2.loc[idx_tr, ['bill_length_mm', 'bill_leng_2', 'bill_leng_3', 'bill_leng_4', 'bill_leng_5', 'bill_leng_6', 'bill_leng_7', 'bill_leng_8', 'bill_leng_9', 'bill_leng_10']]
X_deg10_te = penguins2.loc[idx_te, ['bill_length_mm', 'bill_leng_2', 'bill_leng_3', 'bill_leng_4', 'bill_leng_5', 'bill_leng_6', 'bill_leng_7', 'bill_leng_8', 'bill_leng_9', 'bill_leng_10']]
y_train_deg10 = penguins2.loc[idx_tr, ['bill_depth_mm']]
y_test_deg10 = penguins2.loc[idx_te, ['bill_depth_mm']]
```

```python
# Degree 10 Polynomial Model
deg10_model = LinearRegression()
deg10_model.fit(X_deg10_tr, y_train_deg10)
y_deg10_pred = deg10_model.predict(X_deg10_te)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a scatter plot of the original data
plt.figure(figsize=(10, 6))
plt.scatter(X_test['bill_length_mm'], y_test, label='Actual Data', alpha=0.5)

# Sort the test data for plotting smooth curves
X_test_sorted = np.sort(X_test['bill_length_mm'].values)

# Convert sorted test data to DataFrames with appropriate column names
X_test_sorted_lin = pd.DataFrame(X_test_sorted, columns=['bill_length_mm'])

X_test_sorted_sq = pd.DataFrame(np.c_[X_test_sorted, X_test_sorted**2], columns=['bill_length_mm', 'bill_leng_2'])

X_test_sorted_cub = pd.DataFrame(np.c_[X_test_sorted, X_test_sorted**2, X_test_sorted**3], columns=['bill_length_mm', 'bill_leng_2', 'bill_leng_3'])

X_test_sorted_deg10 = pd.DataFrame(np.c_[X_test_sorted, X_test_sorted**2, X_test_sorted**3, X_test_sorted**4, X_test_sorted**5, X_test_sorted**6, X_test_sorted**7, X_test_sorted**8, X_test_sorted**9, X_test_sorted**10], columns=['bill_length_mm', 'bill_leng_2', 'bill_leng_3', 'bill_leng_4', 'bill_leng_5', 'bill_leng_6', 'bill_leng_7', 'bill_leng_8', 'bill_leng_9', 'bill_leng_10'])


# Predict using each model on the sorted test data
y_lin_pred_sorted = lin_model.predict(X_test_sorted_lin)
y_sq_pred_sorted = sq_model.predict(X_test_sorted_sq)
y_cub_pred_sorted = cub_model.predict(X_test_sorted_cub)
y_deg10_pred_sorted = deg10_model.predict(X_test_sorted_deg10)


# Plot the fitted models
plt.plot(X_test_sorted, y_lin_pred_sorted, label='Linear Model', color='red')
plt.plot(X_test_sorted, y_sq_pred_sorted, label='Quadratic Model', color='green')
plt.plot(X_test_sorted, y_cub_pred_sorted, label='Cubic Model', color='purple')
plt.plot(X_test_sorted, y_deg10_pred_sorted, label='Degree 10 Model', color='orange')


plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Fitted Models on Penguin Data')
plt.legend()
plt.grid(True)
plt.show()
```
